#Mutual Exclusivity Analysis

# Binary Matrix Construction

```python
import autogluon

import pandas as pd
import scanpy as sc

# Import the data
datamutations = pd.read_csv(DATA / "data_mutationslaml.txt", sep='\t')

# Get unique patients and genes
patients = datamutations['Tumor_Sample_Barcode'].unique()[0:]
numpatients = len(patients)
genes = datamutations['Hugo_Symbol'].unique()[0:]
numgenes = len(genes)

# Create a binary matrix
binarymatrix = pd.DataFrame(0, index=patients, columns=genes)

# Populate the binary matrix
for i in range(1, len(datamutations)):
    gene = datamutations.loc[i, 'Hugo_Symbol']
    patient = datamutations.loc[i, 'Tumor_Sample_Barcode']

    if gene in genes and patient in patients:
        binarymatrix.loc[patient, gene] = 1
```

# First Round of Fisher's Exact Tests

```python
import numpy as np
from scipy.stats import fisher_exact
from statsmodels.stats.multitest import multipletests
import time

# Fisher's exact tests for each pair of columns
start_time = time.time()

numgenes = binarymatrix.shape[1]

# Initialize result matrices
pvals = np.zeros((numgenes, numgenes))
odds = np.zeros((numgenes, numgenes))

# Loop through every pair of genes
for i in range(numgenes):
    for j in range(numgenes):
        if i == j:
            pvals[i, j] = 1  # for the same gene, p-value is 1
        else:
            contingencytable = np.array([
                [np.sum(binarymatrix[:, i] & ~binarymatrix[:, j]), np.sum(binarymatrix[:, i] & binarymatrix[:, j])],
                [np.sum(~binarymatrix[:, i] & ~binarymatrix[:, j]), np.sum(~binarymatrix[:, i] & binarymatrix[:, j])]
            ])
            # Fisher's exact test for each pair
            oddsratio, P = fisher_exact(contingencytable, alternative='two-sided')
            # Add results to result matrices
            pvals[i, j] = P
            odds[i, j] = np.log2(oddsratio)

# Calculate FDR corrected q-values using BH method
flat_pvals = pvals.flatten()
rejected, qvals, _, _ = multipletests(flat_pvals, alpha=0.05, method='fdr_bh')

qvals = qvals.reshape((numgenes, numgenes))

end_time = time.time()
elapsed_time = end_time - start_time

print(f"Time taken: {elapsed_time} seconds")
```

# First Greedy Algoritm

```python
import pandas as pd
import numpy as np

# Create DataFrame from the binary matrix and patients
binarymatrix_df = pd.DataFrame(binarymatrix, columns=genes, index=patients)

# Find the index of the "TP53" gene
tp53ind = genes.index("TP53")

# Initialize lists to store results
sigpvals = []
indices = []
me = [tp53ind]

# Find mutually exclusive genes (odds > 0)
for i in range(numgenes):
    if odds[tp53ind, i] > 0:
        sigpvals.append(pvals[tp53ind, i])
        indices.append(i)

# Find the index with the minimum p-value
ind = np.argmin(sigpvals)

# Store p-value and index
p = 1 if not sigpvals else sigpvals[ind]
pvalues = [1, p]

# Store indices
me.extend([indices[ind]])

# Display results
print(me)
print(np.array(genes)[me])
print(pvalues)

# Write results to a text file
with open('lamlresults.txt', 'w') as fid:
    fid.write(' '.join(map(str, np.array(genes)[me])) + '\n')
    fid.write(' '.join(map(str, me)) + '\n')
    fid.write(' '.join(map(str, pvalues)) + '\n\n')

# Update binary data matrix
binarymatrix_df['Module'] = binarymatrix_df.iloc[:, me].sum(axis=1)
binarymatrix_df.loc[binarymatrix_df['Module'] > 1, 'Module'] = 1
binarymatrix_df = binarymatrix_df.drop(columns=genes[me])

# Convert the DataFrame back to a NumPy array
binarymatrix = binarymatrix_df.to_numpy()
```
